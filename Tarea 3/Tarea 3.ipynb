```python
############################### DIAGRAMA H-R ###################################

new_data.insert(loc=1, column='Distancia', value='')
```

```python
new_data.head()
```

```python
new_data.dropna(inplace=True)
new_data.isna().sum()
```

```python
new_data['Distancia'] = (1/(new_data['Plx']*10**-3))
```

```python
new_data
```

```python
new_data['AbsMag'] = (new_data.Vmag - 5 * numpy.log10(new_data.Distancia) + 5)
new_data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,12))

ax.set_xlim(-0.5, 2.5)
ax.set_ylim(15, -10)
ax.grid()
ax.set_title('Diagrama H-R -> Catalogo Hipparcos')

ax.title.set_fontsize(20)
ax.set_xlabel('ndice de color B-V')
ax.set_ylabel('Magnitud absoluta')
plt.scatter(new_data['BV'], new_data['AbsMag'], s=1,color="darkred")
plt.savefig('hr.png')
```
